```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 
# predict_result_0_5_shortest_path = pd.read_csv('predict_result_0_5_shortest_path.csv') 
predict_result_0_shortest_path = pd.read_csv('predict_result_0_shortest_path.csv')
predict_result_1_shortest_path = pd.read_csv('predict_result_1_shortest_path.csv')
predict_result_optimized = pd.read_csv('predict_result_optimized.csv')

# 
columns_to_plot = [
    'matching_prob', 'ride_distance', 'detour_distance',
    'shared_distance', 'saved_distance', 'ride_distance_for_taker',
    'detour_distance_for_taker', 'shared_distance_for_taker',
    'saved_distance_for_taker', 
    'detour_distance_for_seeker', 'shared_distance_for_seeker',
    'saved_distance_for_seeker'
]

# predict_result_0_5_shortest_path = predict_result_0_5_shortest_path[columns_to_plot]
predict_result_0_shortest_path = predict_result_0_shortest_path[columns_to_plot]
predict_result_1_shortest_path = predict_result_1_shortest_path[columns_to_plot]
predict_result_optimized = predict_result_optimized[columns_to_plot]

# 
# predict_result_0_5_shortest_path['Dataset'] = '0.5 Shortest Path'
predict_result_0_shortest_path['Dataset'] = '0 Shortest Path'
predict_result_1_shortest_path['Dataset'] = '1 Shortest Path'
predict_result_optimized['Dataset'] = 'Optimized'

# 
df = pd.concat([
    # predict_result_0_5_shortest_path,
    predict_result_0_shortest_path,
    predict_result_1_shortest_path,
    predict_result_optimized
])

# 
df = df[columns_to_plot + ['Dataset']]

# 
df_melted = df.melt(id_vars=['Dataset'], var_name='Feature', value_name='Value')

# 
for feature in columns_to_plot:
    plt.figure(figsize=(8, 6))  # 
    sns.boxplot(x='Dataset', y='Value', data=df_melted[df_melted['Feature'] == feature])
    plt.title(f'Boxplot of {feature}')
    plt.xticks(rotation=20)  #  x 
    plt.xlabel('Dataset')
    plt.ylabel('Value')
    plt.grid(axis='y', linestyle='--', alpha=0.7)  # 
    plt.show()
```
