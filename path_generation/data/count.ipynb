```python
import pandas as pd
import os

# 确认当前目录下的文件
print(os.listdir())
files = os.listdir('./orders')

results = []

for file in files:
    
    # 读取数据
    data = pd.read_csv('./orders/'+ file)

    # 将 'departure_time' 转换为 datetime 类型
    data['departure_time'] = pd.to_datetime(data['dwv_order_make_haikou_1.departure_time'])

    # 设置时间间隔，例如每15分钟
    interval = pd.Timedelta(minutes=15)

    # 按照 'departure_time' 和 'O_location' 进行分组，并统计每个间隔内的订单数量
    order_counts = data.groupby([pd.Grouper(key='departure_time', freq=interval), 'O_location']).size().reset_index()
    order_counts.columns = ['departure_time', 'O_Location', 'OrderCount']
    order_counts['interval'] = pd.to_datetime(order_counts['departure_time']).dt.time
    results.append(order_counts)

# 合并所有结果
df = pd.concat(results, ignore_index=True)
df.head()
```

```python
tmp = df.groupby(['interval','O_Location'])['OrderCount'].sum().reset_index()
tmp
```
