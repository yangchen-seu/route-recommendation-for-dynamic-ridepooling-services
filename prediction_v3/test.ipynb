```python
import networkx as nx
from heapq import heappush, heappop

def yen_k_shortest_paths(G, source, target, k, weight='weight'):
    """
    使用 Yen's Algorithm 计算图 G 中从 source 到 target 的 K 最短路径。
    :param G: NetworkX 图对象
    :param source: 起点
    :param target: 终点
    :param k: 需要的最短路径数量
    :param weight: 边的权重属性名
    :return: K 最短路径列表及其长度
    """
    # 计算最短路径作为第一个路径
    paths = []
    distances = []
    try:
        first_path = nx.shortest_path(G, source, target, weight=weight)
        first_distance = sum(G[u][v][weight] for u, v in zip(first_path[:-1], first_path[1:]))
    except nx.NetworkXNoPath:
        return paths, distances  # 如果无路径，直接返回空列表
    
    paths.append(first_path)
    distances.append(first_distance)

    # 用于存储替代路径的优先队列
    potential_paths = []

    for k_index in range(1, k):
        # 上一个路径的根路径处理
        for i in range(len(paths[-1]) - 1):
            spur_node = paths[-1][i]
            root_path = paths[-1][:i + 1]

            # 临时移除 root_path 的边，防止重复计算
            removed_edges = []
            for path in paths:
                if len(path) > i and path[:i + 1] == root_path:
                    edge = (path[i], path[i + 1])
                    if G.has_edge(*edge):
                        removed_edges.append((edge, G[edge[0]][edge[1]][weight]))
                        G.remove_edge(*edge)

            # 计算从 spur_node 到目标的最短路径
            try:
                spur_path = nx.shortest_path(G, spur_node, target, weight=weight)
                spur_distance = sum(G[u][v][weight] for u, v in zip(spur_path[:-1], spur_path[1:]))
                total_path = root_path[:-1] + spur_path
                total_distance = sum(G[u][v][weight] for u, v in zip(total_path[:-1], total_path[1:]))
                heappush(potential_paths, (total_distance, total_path))
            except nx.NetworkXNoPath:
                pass  # 无路径时跳过

            # 恢复移除的边
            for edge, w in removed_edges:
                G.add_edge(edge[0], edge[1], weight=w)

        # 如果没有候选路径，退出
        if not potential_paths:
            break

        # 取出权重最小的候选路径
        distance, path = heappop(potential_paths)
        paths.append(path)
        distances.append(distance)

    return paths, distances



# 计算从source 到target 的 k 条最短路径
source = 0
target = 10
k = 3
paths, distances = yen_k_shortest_paths(G, source, target, k)

print("K 最短路径:")
for i, (path, distance) in enumerate(zip(paths, distances), 1):
    print(f"第 {i} 条路径: {path}, 距离: {distance}")
```

```python

import networkx as nx

def construct_subgraph(G, os, od_star, alpha = 1.5):
    """
    构建子图 D = (V_D, E_D)
    :param G: 完整的路网图 (NetworkX DiGraph)
    :param os: 起点 (订单的 pickup 点)
    :param od_star: 目标点 (订单的 drop-off 点)
    :return: 子图 D (NetworkX DiGraph)
    """
    # 计算最短路径长度
    sp_os_to_od_star = nx.shortest_path_length(G, source=os, target=od_star, weight='weight')
    print('最短路',sp_os_to_od_star)
    length, path = nx.single_source_dijkstra(G, source=os, target=od_star, weight='weight')

    print("最短路径长度:", length)
    print("最短路径:", path)

    # # 确定子图节点集 V_D
    # V_D = {
    #     v for v in G.nodes
    #     if nx.shortest_path_length(G, source=v, target=od_star, weight='weight') <= sp_os_to_od_star
    # }

    # 确定子图节点集 V_D
    V_D = {
        v for v in G.nodes
        if nx.shortest_path_length(G, source=os, target=v, weight='weight') + nx.shortest_path_length(G, source=v, target=od_star, weight='weight') <= alpha * sp_os_to_od_star
    }
    
    # 确定子图边集 E_D
    E_D = [
        (u, v) for u, v in G.edges
        if u in V_D and v in V_D and
           nx.shortest_path_length(G, source=v, target=od_star, weight='weight') <=
           nx.shortest_path_length(G, source=u, target=od_star, weight='weight')
    ]
    print('LEN(ed)',len(E_D),E_D)
    for u, v in G.edges:
        if u in V_D and v in V_D:
            if nx.shortest_path_length(G, source=u, target=od_star, weight='weight') <= nx.shortest_path_length(G, source=v, target=od_star, weight='weight'):
               E_D.append((v,u))
    print('LEN(ed)',len(E_D),E_D)
    
    # 构建子图
    D = nx.DiGraph()
    D.add_nodes_from(V_D)
    # 添加边和边属性
    for u, v in E_D:
        if G.has_edge(u, v):
            attributes = G.get_edge_data(u, v)  # 获取原图中的边属性
            D.add_edge(u, v, **attributes)
    
    # 计算拓扑排序
    topological_order = list(nx.topological_sort(D))

    # 输出节点及其拓扑排序
    topological_order_dict = {node: i+1 for i, node in enumerate(topological_order)}  # 排序从1开始

    # 如果需要将拓扑顺序作为属性存储到图中
    nx.set_node_attributes(D, topological_order_dict, name="topological_order")


    return D


    
# 使用起点和目标点构建子图
os = 0
od_star = 10
D = construct_subgraph(G, os, od_star)
print(D.nodes)
print('edges',D.edges)


# 提取节点及其拓扑排序
data = [(node, attributes["topological_order"]) for node, attributes in D.nodes(data=True)]

# 转为 DataFrame
df = pd.DataFrame(data, columns=["节点", "拓扑排序"])
df = df.sort_values(by="拓扑排序", ascending=False)
print(df)

def remove_nodes_after_a(G, a):
    # 确保图是有向无环图
    if not nx.is_directed_acyclic_graph(G):
        raise ValueError("The graph is not a DAG.")

    # 获取拓扑排序
    topo_sort = list(nx.topological_sort(G))

    # 找到节点 a 的位置
    if a not in topo_sort:
        raise ValueError(f"Node {a} is not in the graph.")

    a_index = topo_sort.index(a)

    # 获取节点 a 之后的所有节点
    nodes_to_remove = topo_sort[a_index + 1:]

    # 从图中移除这些节点
    G.remove_nodes_from(nodes_to_remove)

    return G


length, path = nx.single_source_dijkstra(D, source=os, target=od_star, weight='weight')

print("最短路径长度:", length)
print("最短路径:", path)

D_ = remove_nodes_after_a(D, od_star)
# 提取节点及其拓扑排序
data = [(node, attributes["topological_order"]) for node, attributes in D_.nodes(data=True)]
print('D_.nodes',D_.nodes,)
topological_order_dict = nx.get_node_attributes(D_, "topological_order")
print(topological_order_dict)

length, path = nx.single_source_dijkstra(D_, source=os, target=od_star, weight='weight')

print("最短路径长度:", length)
print("最短路径:", path)

# 转为 DataFrame
df = pd.DataFrame(data, columns=["节点", "拓扑排序"])
df = df.sort_values(by="拓扑排序", ascending=False)
print(df)
```

```python
def dp_path_finding(G, os, od_star, l_min, bins=100):
    """
    Dynamic Programming algorithm for finding the optimal path in a DAG.

    :param G: The original graph (NetworkX DiGraph).
    :param os: Starting node.
    :param od_star: Destination node.
    :param alpha: Detour threshold.
    :param bins: Number of bins for discretization.
    :return: Optimal path and its expected number of compatible passengers.
    """
    # Step 1: Construct the DAG and compute l_min
    DAG = G
    
    bin_width = l_min / bins

    # Step 2: Compute topological order
    topo_order = sorted(nx.get_node_attributes(DAG, "topological_order").items(), key=lambda x: x[1])
    topo_order = [node for node, _ in topo_order]  # Extract nodes in topological order
    
    # Step 3: Initialize DP table
    dp = {v: [-np.inf] * (bins + 1) for v in DAG.nodes}
    backtrack = {v: [-1] * (bins + 1) for v in DAG.nodes}
    dp[os][0] = 0  # Starting point

    print('topo_order',topo_order)
    
    print('DAG.nodes',DAG.nodes)
    
    # Step 4: Process nodes in topological order
    for v in topo_order:
        for b in range(bins + 1):
            if dp[v][b] == -np.inf:
                continue

            # Update neighbors
            for u, data in DAG[v].items():
                print('u',u,'data',data)
                weight = data['weight']
                delta_b = int(np.ceil(weight / bin_width))

                if b + delta_b <= bins:
                    new_b = b + delta_b
                    expected_passengers = data.get('expected_passengers', 0)  # Default 0 if not provided

                    if dp[v][b] + expected_passengers > dp[u][new_b]:
                        print('dp[v][b] ',dp[v][b] ,'expected_passengers',expected_passengers, 'v',v, 'b',b)
                        dp[u][new_b] = dp[v][b] + expected_passengers
                        backtrack[u][new_b] = v

    # Step 5: Find the optimal path
    best_score = -np.inf
    best_bin = -1

    for b in range(bins + 1):
        if dp[od_star][b] > best_score:
            best_score = dp[od_star][b]
            print('best_score',best_score,'b',b)
            best_bin = b

    # Step 6: Backtrack to get the optimal path
    path = []
    current_node = od_star
    current_bin = best_bin

    while current_node != -1:
        path.append(current_node)
        current_node = backtrack[current_node][current_bin]
        current_bin -= int(np.ceil(G[current_node][path[-1]]['weight'] / bin_width)) if current_node != -1 else 0

    path.reverse()
    return path, best_score


os = 0
od_star = 10
alpha = 1.5  # Detour threshold (not used in this implementation)
length, path = nx.single_source_dijkstra(D_, source=os, target=od_star, weight='weight')

l_min = length * alpha
path, score = dp_path_finding(D_, os, od_star, l_min, bins=100)
print("Optimal Path:", path)
print("Maximum Expected Passengers:", score)

```

```python
import networkx as nx

def find_max_weight_path_dag_with_length(G, source, target):
    # 获取拓扑排序
    topological_order = list(nx.topological_sort(G))
    
    # 初始化最大权路径字典，起点权值为0，其他为负无穷
    max_weight = {node: float('-inf') for node in G.nodes}
    max_weight[source] = 0
    
    # 辅助字典存储前驱节点以便回溯路径
    predecessor = {node: None for node in G.nodes}
    
    # 遍历拓扑排序后的节点
    for u in topological_order:
        if max_weight[u] == float('-inf'):
            continue  # 如果节点不可达，跳过
        for v, attr in G[u].items():
            weight = attr['passenger_number']  # 当前用于优化的权值
            if max_weight[u] + weight > max_weight[v]:
                max_weight[v] = max_weight[u] + weight
                predecessor[v] = u
    
    # 如果目标节点不可达
    if max_weight[target] == float('-inf'):
        return None, float('-inf'), 0  # 不可达时返回路径、权值和权重长度为0
    
    # 回溯路径并计算路径长度（以 weight 为权重）
    path = []
    total_weight_length = 0  # 路径的权重长度
    current = target
    while current is not None:
        if predecessor[current] is not None:  # 累加边的权重
            total_weight_length += G[predecessor[current]][current]['weight']
        path.append(current)
        current = predecessor[current]
    path.reverse()
    
    return path, max_weight[target], total_weight_length



source = 0
target = 10

# 调用函数
path, max_passengers, path_weight_length = find_max_weight_path_dag_with_length(D_, source, target)

print("最大权路径:", path)
print("路径权值（乘客数）:", max_passengers)
print("路径长度（权重累积）:", path_weight_length)
```
